```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head()
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

```python
age=df.Age
all_devs=df.All_Devs
py=df.Python
js=df.JavaScript
```

```python
plt.plot(age, py, label='Python', linewidth=3, marker= 'o', color="#2214BD")
plt.plot(age, all_devs, label='All Devs',marker="o",color="#C50808")
plt.xlabel("Age",fontsize=15)
plt.ylabel("Languages",fontsize=15)
plt.title("This is the graph",fontsize=20,color="#DE1AAD")
plt.grid()
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
