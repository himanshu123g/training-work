```python
df.drop(df[df["Rating"]>5].index,inplace=True)
```

```python
df.Rating.fillna(df.Rating.mean(),inplace=True)
```

```python
df.Type.fillna("Free",inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.info()
```

```python
df.Size.unique()
```

```python
df.drop(df[df.Size =="Varies with device"].index,inplace=True)
```

```python
df.Size.unique()
```

```python
df.Size=df.Size.str.replace("k","")
```

```python
df.Size.unique()
```

```python
df["New_size"]=df.Size.apply(lambda x: int(x)/1024 if str(x).isdigit() else x)
```

```python
df["New_size"]=df.New_size.str.replace("M","")
```

```python
df["New_size"]=df.New_size.astype("float")
```

```python
df.dropna(inplace=True)
```
