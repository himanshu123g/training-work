```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("data.csv")
```

```python
df.head()
```

```python
age=df.Age
all_devs=df.All_Devs
py=df.Python
js=df.JavaScript
```

```python
print(plt.style.available)
```

```python
plt.style.use("dark_background")
plt.plot(age,py,label="Python",marker="o",linewidth=3,color="#D912B4")
plt.plot(age,js,label="Javascript",color="#F6F6F6",linewidth=3,linestyle="--",marker="o")
plt.title("This is a graph",fontsize=12)
plt.xlabel("Ages",fontsize=10)
plt.ylabel("Subjects",fontsize=10)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
