```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("spam.csv")
```

```python
df["Output"]=df.Category.map({"spam":1,"ham":0})
```

```python
df.head()
```

```python
df.shape
```

```python
df.Message[2]
```

```python
x=df.Message
y=df.Output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
type(x_train)
```

```python
x_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
x_train_cv=cv.fit_transform(x_train)
```

```python
x_train_cv
```

```python
cv.get_feature_names_out()[5000]
```

```python
cv.vocabulary_
```

```python
x_train_np=x_train_cv.toarray()
```

```python
x_train_np
```

```python
x_train_np.shape
```

```python
np.where(x_train_np[5]!=0)
```

```python
x_train[5]
```

```python
y_train[5]
```

```python
from sklearn.naive_bayes import MultinomialNB
MB=MultinomialNB()
```

```python
model=MB.fit(x_train_np,y_train)
model
```

```python
x_test_cv=cv.transform(x_test)
```

```python
x_test_np=x_test_cv.toarray()
```

```python
y_pred=model.predict(x_test_np)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
