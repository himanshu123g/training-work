```python
record = pd.DataFrame({
  'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
  'Age': [21, 19, 20, 18, 17, 21],
  'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],'Percentage': [88, 92, 95, 70, 65, 78]})
```

```python
record
```

```python
record.loc[(record["Stream"]=="Math") | (record["Stream"]=="Commerce")]
```

```python
record.loc[record["Stream"]!="Science"]
```

```python
record
```

```python
record.loc[(record["Stream"]=="Science") & (record["Age"]<20)]
```

```python
record.loc[(record["Stream"]=="Science") | (record["Stream"]=="Math"), "Percentage"].sum()
```

```python
record.loc[record["Stream"]=="Science","Percentage"].max()
```
