```python
count_val=int(input("Enter the number of values :- "))
list_1=[]
for i in range(count_val):
    element=int(input("Enter the element value :- "))
    list_1.append(element)
print(list_1)
```

```python
import pandas as pd
series=pd.Series(list_1)
sum=0
for i in series:
    sum+=i
print("The average of both of numbers will be :- ",sum/count_val)
```

```python
list_squared=[]
print("The values are after the square are :- ")
for i in series:
    squared_values=(i-(sum/count_val))*((i-(sum/count_val)))
    list_squared.append(squared_values)
print(list_squared)
    
```

```python
import math as m
squared_mean=[]
sum_squared=0
for j in list_squared:
    sum_squared+=j
numb=sum_squared/count_val
print(numb)
square_root=m.sqrt(numb)
print("The standard deviation of the number is :- ",square_root)
```
