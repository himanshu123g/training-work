```python
elements=int(input("Enter the number of elements you want a list of :- "))
i=0
list_1=[]
for i in range(elements):
    value=int(input("Enter the values :- "))
    list_1.append(value)
print("The values in the list are :- ",list_1)
```

```python
square_index_result=[]
for j in list_1:
    square_index=j*j
    square_index_result.append(square_index)
print(square_index_result)
```
