```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv("Housing.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df.mainroad.unique()
```

```python
df["mainroad"]=df.mainroad.map({"yes":1,"no":0})
```

```python
df.guestroom.unique()
```

```python
df["guestroom"]=df.guestroom.map({"yes":1,"no":0})
df["basement"]=df.basement.map({"yes":1,"no":0})
df["hotwaterheating"]=df.hotwaterheating.map({"yes":1,"no":0})
df["airconditioning"]=df.airconditioning.map({"yes":1,"no":0})
df["prefarea"]=df.prefarea.map({"yes":1,"no":0})
```

```python
df.furnishingstatus.unique()
```

```python
x=df.drop(columns="price")
```

```python
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop="first",sparse_output=False)
```

```python
x_train_ohe=ohe.fit_transform(x_train[["furnishingstatus"]])
x_test_ohe=ohe.transform(x_test[["furnishingstatus"]])
```

```python
feature=ohe.get_feature_names_out(["furnishingstatus"])
```

```python
x_train_ohe=pd.DataFrame(x_train_ohe, columns=feature)
x_test_ohe=pd.DataFrame(x_test_ohe, columns=feature)
```

```python
x_train=x_train.reset_index(drop=True)
x_test=x_test.reset_index(drop=True)
```

```python
x_train=x_train.drop(columns="furnishingstatus")
x_test=x_test.drop(columns="furnishingstatus")
```

```python
x_train=pd.concat([x_train,x_train_ohe],axis=1)
x_test=pd.concat([x_test,x_test_ohe],axis=1)
```

```python
x_train[feature]=x_train_ohe
x_test[feature]=x_test_ohe
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```

```python
import pickle
pickle.dump(lr,open("model.pkl","wb"))
pickle.dump(sc,open("scalar.pkl","wb"))
pickle.dump(ohe,open("encoder.pkl","wb"))
```
