```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.read_csv("car_purchasing.csv", encoding="latin1")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["customer name","country","customer e-mail"],inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns="car purchase amount")
y=df["car purchase amount"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
x.shape
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential([
    keras.layers.Dense(25,activation="relu",input_shape=(5,)),
    keras.layers.Dense(25,activation="relu"),
    keras.layers.Dense(1)
])
```

```python
model.compile(optimizer="Adam",metrics=["mae"],loss="mean_squared_error")
```

```python
model.fit(x=x_train,y=y_train,epochs=50)
```

```python
y_pred=model.predict(x_test)
```

```python
model.evaluate(x_test,y_test)
```

```python
print(y_test[:5])
print(y_pred[:5])
```

```python
model.save("Carmodel.h5")
```
