```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.layers import Embedding, Dense, SimpleRNN
from keras.models import Sequential
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
texts = [
    "I love this product",
    "This is the worst thing ever",
    "Absolutely fantastic!",
    "Not good at all",
    "Amazing experience",
    "Terrible, I hate it",
]
labels = [1, 0, 1, 0, 1, 0]
```

```python
Tokenizer=Tokenizer(num_words=1000)
```

```python
Tokenizer.fit_on_texts(texts)
```

```python
sequences=Tokenizer.texts_to_sequences(texts)
sequences
```

```python
length=[]
for i in sequences:
    length.append(len(i))

print(max(length))
```

```python
padded_data=pad_sequences(sequences,padding="post",maxlen=10)
```

```python
padded_data
```

```python
x_train,y_train=padded_data[0:4],np.array(labels[0:4])
x_test,y_test=padded_data[4:],np.array(labels[4:])
```

```python
model=Sequential([
    Embedding(input_dim=1000,output_dim=16,input_length=10),
    SimpleRNN(16),
    keras.layers.Dense(1,activation="sigmoid")
])
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=20,batch_size=2,verbose=1)
```

```python
loss,accuracy=model.evaluate(x_test,y_test)
```

```python
print(loss)
```

```python
print(accuracy)
```

```python
new_text=["I like this product","Not good at all","you are not good"]
new_sequences=Tokenizer.texts_to_sequences(new_text)
```

```python
new_padded=pad_sequences(new_sequences,padding="post",maxlen=10)
```
